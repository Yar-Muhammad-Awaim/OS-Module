### `mkdirs()` method

__Difference between `mkdir()` and `mkdirs()` method.__

* mkdir method creates a directory and if it already exists, this method throws `FileExistsError` error. The main difference is that if we are creating `foo\bar` and both `foo` and `bar` doesn't exist. `mkdir()` method will fail here. This would be handled by `mkdirs()` method. It will create both `foo` and `bar` directories.

```python
print(f"You are on {os.getcwd()}.")
os.makedirs("./foo/baz")
print(f"The current working directory has following files: {os.listdir()}")
```

```python
print("The foo directory has following files: ")
os.listdir("./foo")
```

### Deleting a File

In order to delete a file we need to use `remove()` method of os module.
* __NOTE__ This method can only delete files and `OSError` will be raised if the specified path is a directory and not a file.

```python
os.system("notepad os-test.txt")
```

```python
os.remove("./os-test.txt")
```
